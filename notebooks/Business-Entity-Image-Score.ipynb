```python
score_df = pd.read_csv('/Users/stjepankusenic/POLAR_WEBE/data/processed/POLAR-GoogleNews-score-antonyms-inter.csv')
score_df.head()
```

```python
corporate=['corrupthonest','environmentpollution','sustainableunsustainable','bigsmall','harmfulhelpful','irrelevantrelevant']
score_df['corporate_score']= score_df[corporate].mean(axis=1)
product=['affordableexpensive','qualityworthless','commonlimited','usefulluseless','fewmany','demandexcess','valuableworthless','currentoutdated','complexsimple']
score_df['product_score']=score_df[product].mean(axis=1)
price=['cheappricy','highlow','fairunfair','partialtotal','reasonableunreasonable','accurateinaccurate','clearunclear']
score_df['price_score']=score_df[price].mean(axis=1)
vendor=['onlinestore','crowdedempty','cleandirty','conventionalinnovative','chaoticorganized','brightdull','convenientinconvenient','attractiveunattractive']
score_df['vendor_score']=score_df[vendor].mean(axis=1)
service=['friendlyrude','helpfulunhelpful','badgood','calmhectic','directvague','calmnesspressure','fastslow','responsiveunresponsive']
score_df['service_score']=score_df[service].mean(axis=1)
workforce=['poorwealthy','intelligentstupid','oldyoung','beginnerexperienced','managerworker','renownedunknown','successfulunsuccessful','reliableunreliable','enoughinsufficient']
score_df['workforce_score']=score_df[workforce].mean(axis=1)
marketing=['frequentinfrequent','boringinteresting','flashysimple','meaningfullmeaningless','happyserious','effectiveineffective','informativeuninformative','falsetrue']
score_df['marketing_score']=score_df[marketing].mean(axis=1)
```
