```python
#Import all the libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline
```

```python
#Read all the data files
gnews = pd.read_csv('../data/processed/POLAR-Google-news-antonyms-inter.csv')
reddit = pd.read_csv('../data/processed/POLAR-reddit-antonyms-inter.csv')
wiki = pd.read_csv('../data/processed/POLAR-wiki-antonyms-inter.csv')
twitter = pd.read_csv('../data/processed/POLAR-twitter-antonyms-inter.csv')
```

```python
#Change the Unnamed column to Company Name
reddit = reddit.rename(columns = {'Unnamed: 0' : 'Company Name'})
wiki = wiki.rename(columns = {'Unnamed: 0' : 'Company Name'})
twitter = twitter.rename(columns = {'Unnamed: 0' : 'Company Name'})
gnews = gnews.rename(columns = {'Unnamed: 0' : 'Company Name'})

#Read the fortune_500 list to get the countries
fortune_500 = pd.read_csv('../data/processed/fortune_500.csv', encoding= 'unicode_escape')
```

```python
#ADD Countries
fortune_500 = fortune_500[['Company Name', 'Country']]
fortune_500['Company Name'] = fortune_500['Company Name'].str.lower()
fortune_500['Company Name'] = fortune_500['Company Name'].drop_duplicates().str.split().str[0]
```

```python
#Merge the files
reddit = pd.merge(reddit, fortune_500, on='Company Name', how='left')
wiki = pd.merge(wiki, fortune_500, on='Company Name', how='left')
twitter = pd.merge(twitter, fortune_500, on='Company Name', how='left')
gnews = pd.merge(gnews, fortune_500, on='Company Name', how='left')
```

```python
reddit["Source"] = "reddit"
wiki["Source"] = "wiki"
twitter["Source"] = "twitter"
gnews["Source"] = "gnews"
```

```python
#Get common companies in all the embeddings
list1 = []
list1 = gnews['Company Name']
reddit = reddit[reddit['Company Name'].isin(list1)]
wiki = wiki[wiki['Company Name'].isin(list1)]
twitter = twitter[twitter['Company Name'].isin(list1)]
```

```python
reddit['Country'] = reddit['Country'].astype(str) + '_reddit'
wiki['Country'] = wiki['Country'].astype(str) + '_wiki'
twitter['Country'] = twitter['Country'].astype(str) + '_twitter'
gnews['Country'] = gnews['Country'].astype(str) + '_gnews'
```

```python
reddit['Country'] = reddit['Country'].astype(str)
wiki['Country'] = wiki['Country'].astype(str)
twitter['Country'] = twitter['Country'].astype(str)
gnews['Country'] = gnews['Country'].astype(str)
```

```python
reddit_country = reddit.groupby('Country').mean()
reddit_country = reddit_country.reset_index()
wiki_country = wiki.groupby('Country').mean()
wiki_country = wiki_country.reset_index()
twitter_country = twitter.groupby('Country').mean()
twitter_country = twitter_country.reset_index()
gnews_country = gnews.groupby('Country').mean()
gnews_country = gnews_country.reset_index()
```

```python
combination_country = pd.concat([reddit_country, wiki_country, twitter_country, gnews_country])
combination_country.reset_index()
```

```python
reddit['Company Name'] = reddit['Company Name'].astype(str) + '_reddit'
wiki['Company Name'] = wiki['Company Name'].astype(str) + '_wiki'
twitter['Company Name'] = twitter['Company Name'].astype(str) + '_twitter'
gnews['Company Name'] = gnews['Company Name'].astype(str) + '_gnews'
```

```python
reddit['Company Name'] = reddit['Company Name'].astype(str)
wiki['Company Name'] = wiki['Company Name'].astype(str)
twitter['Company Name'] = twitter['Company Name'].astype(str)
gnews['Company Name'] = gnews['Company Name'].astype(str)
```

```python
combination = pd.concat([reddit, wiki, twitter, gnews], ignore_index=True)
```

```python
combination = combination.sort_values(['Company Name'], ascending=True)
```

```python
from scipy.stats import ttest_ind

ttest_ind(usa['marketing secret'], non_usa['marketing secret'])
```

```python
for i in combination.groupby("Source").mean().columns:
  s, p = ttest_ind(usa[i], non_usa[i])
  if p <= 0.05:
    print(i, s, p)
```
