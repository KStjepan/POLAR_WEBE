```python
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook as tqdm
import gensim
import sklearn
import nltk
```

```python
# Load Google news vectors
word2vec = gensim.models.KeyedVectors.load_word2vec_format('/Users/stjepankusenic/POLAR_WEBE/data/raw/glove_norm_300.mod',binary=True)
# contains the list of all unique words in pre-trained word2vec vectors
w2v_vocabulary = word2vec.vocab
```

```python
#fortune_data=pd.read_csv('/Users/stjepankusenic/Desktop/Fortune_1000.csv')
#fortune_data.head()
company_data = pd.read_excel(r'/Users/stjepankusenic/POLAR_WEBE/data/raw/Fortune-Global-500-List-2021-Someka-V1F.xlsx')
company_names = company_data['Unnamed: 2'][7:]
display(company_names)
```

```python
#company_names=fortune_data['company']
#display(company_names)
list_of_companies=[]
for name in company_names:
    name1=name.replace(" ","")
    name1=name1.lower()
    list_of_companies.append(name1)
#display(list_of_companies)
len(list_of_companies)
display(list_of_companies)
```

```python
list_of_names=[]
for name in list_of_companies:
    if name in w2v_vocabulary:
        list_of_names.append(name)
display(list_of_names)
```
